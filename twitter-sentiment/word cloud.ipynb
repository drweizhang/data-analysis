```python
import pandas as pd

df = pd.read_excel("output/IVF 2018 -2021 all time w reply.xlsx")

df.shape
```

```python
df['processed_text_rm_standard_sw']

all_words = []
# all_words = list(w for w in df['processed_text_rm_standard_sw'].values)

for text in df['processed_text_rm_standard_sw'].values.tolist():
    #print(type(text))
    ll = text.strip("']['").split("', '")
    all_words = all_words + (ll)

all_words
```

```python
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import re
import numpy as np
import matplotlib.pyplot as plt

def word_cloud(wd_list):
    all_words = ' '.join([text for text in wd_list])
    wordcloud = WordCloud(
        background_color='white',
        stopwords=list(STOPWORDS),
        width=1600,
        height=800,
        random_state=21,
        colormap='jet',
        max_words=50,
        max_font_size=200).generate(all_words)
    plt.figure(figsize=(12, 10))
    plt.axis('off')
    plt.imshow(wordcloud, interpolation="bilinear");

    
word_cloud(all_words)
```
